**Proses Penerjemahan Dataset**

Laporan EKG asli dalam dataset ditulis dalam tiga bahasa utama, yaitu 70,89% dalam bahasa Jerman, 27,9% dalam bahasa Inggris, dan 1,21% dalam bahasa Swedia. Untuk memastikan konsistensi dalam analisis, laporan tersebut diterjemahkan ke dalam bahasa Inggris menggunakan **GoogleTranslator** dari library `deep_translator`. Proses ini diawali dengan deteksi bahasa menggunakan `langdetect`, di mana hanya teks yang bukan dalam bahasa Inggris yang diterjemahkan. Jika teks sudah dalam bahasa Inggris, maka dibiarkan seperti aslinya. Agar proses penerjemahan lebih efisien, digunakan `tqdm` untuk menampilkan progress bar saat menerapkan fungsi penerjemahan ke seluruh dataset. Setelah proses ini selesai, kolom asli **caption** dihapus dan kolom hasil terjemahan **translated_caption** diubah namanya menjadi **caption** untuk mempermudah penggunaan dalam analisis selanjutnya. Selain itu, seluruh teks dalam kolom ini dikonversi menjadi huruf kecil guna menjaga konsistensi format data.

```python
from deep_translator import GoogleTranslator
from langdetect import detect, DetectorFactory
from tqdm import tqdm
import pandas as pd

# Pastikan hasil deteksi bahasa konsisten
DetectorFactory.seed = 0  

def translate_text(text, target_lang='en'):
    try:
        if pd.isna(text) or text.strip() == "":  # Jika teks kosong atau NaN, biarkan apa adanya
            return text
        
        source_lang = detect(text)  # Deteksi bahasa
        if source_lang != target_lang:  # Hanya terjemahkan jika bukan target language
            return GoogleTranslator(source=source_lang, target=target_lang).translate(text)
        return text  # Jika sudah dalam target language, biarkan seperti aslinya
    except Exception as e:
        return text  # Jika gagal deteksi bahasa atau terjadi error, biarkan asli

# Aktifkan tqdm untuk progress bar
tqdm.pandas()

# Buat salinan agar aman
ptbxl = ptbxl.copy()

# Terapkan terjemahan ke kolom baru
ptbxl['translated_caption'] = ptbxl['caption'].astype(str).progress_apply(translate_text)
```

Untuk meningkatkan keterbacaan dan akurasi analisis teks medis, proses standarisasi dilakukan dengan mengganti singkatan dalam kolom caption menggunakan kamus abbreviation_dict. Setiap teks dipecah menjadi kata-kata terpisah, lalu dibandingkan dengan daftar singkatan yang tersedia. Jika kata ditemukan dalam kamus, maka diganti dengan bentuk lengkapnya; jika tidak, tetap dibiarkan seperti aslinya. Setelah proses penggantian selesai, kata-kata tersebut digabungkan kembali menjadi teks lengkap. 

Kode di bawah ini mendefinisikan fungsi replace_abbreviations untuk menggantikan singkatan dalam teks berdasarkan dictionary abbrev_dict. Fungsi ini memecah teks menjadi daftar kata, lalu mengganti setiap kata yang ada dalam dictionary dengan bentuk lengkapnya, dan akhirnya menyatukan kembali kata-kata tersebut menjadi teks yang telah diperbarui. Kemudian, fungsi ini diterapkan pada setiap baris dalam kolom 'caption' pada DataFrame ptbxl menggunakan apply(), sehingga semua singkatan dalam dataset digantikan dengan versi lengkapnya secara otomatis.

```python
import re

def replace_abbreviations(text, abbrev_dict):
    words = text.split()  # Pecah teks jadi list kata
    replaced_words = [abbrev_dict[word.lower()] if word.lower() in abbrev_dict else word for word in words]
    return " ".join(replaced_words)

ptbxl['caption'] = ptbxl['caption'].apply(lambda x: replace_abbreviations(x, abbreviation_dict))
```

```python
import re

# Fungsi untuk menghapus tanda kutip tunggal dan ganda
def remove_quotes(text):
    return re.sub(r"[\"']", "", text)

# Terapkan fungsi ke kolom caption
ptbxl['caption'] = ptbxl['caption'].apply(remove_quotes)
```
